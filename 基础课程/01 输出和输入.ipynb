# Jupyterlab（Jupyter Notebook）介绍
## 两种基本模式
### 命令模式
命令模式将键盘命令与Jupyter Notebook笔记本命令相结合，可以通过键盘不同键的组合运行笔记本的命令。
#### 快捷键
- `Enter`：进入编辑模式
- `Y`：将单元格切换到代码的编辑模式
- `M`：将单元格切换到Markdown的编辑模式
- `A`：在上方插入单元格
- `B`：在下方插入单元格
- `Z`：撤销删除
- `D,D`：删除

### 编辑模式
编辑模式使用户可以在单元格内编辑代码或文档。
#### 快捷键
- `ESC`：退出编辑模式，切换到命令模式
- `Ctrl+Enter`：运行代码
- `Shift+Enter`：运行代码，并在移动到下一个单元格
- `Alt+Enter`：运行代码，并在下方插入单元格

## 保存文档
- Jupyterlab（Jupyter Notebook）会定时保存文档
- 我们也可以通过点击菜单上面的保存图标来保存文档内容和代码
- 未保存的文档，选项卡上面会显示一个实心的圆点。保存后，圆点消失，变成一个“x”的关闭按钮

## 文件夹、笔记本的创建和复制
- 学生可以自行创建新的文件夹，或者笔记本
- 学生自己创建的文件夹或者笔记本，其他学生都无法编辑，只能查看
- 如果需要编辑老师或者其他学生的文件，需要把该文件复制粘贴到自己的文件夹内，然后才可以编辑、保存笔记本

## 注意
在```python_basics```文件夹里面的文件，学生是无法编辑的，所以每次上课的时候，必须做的就是把这节课的笔记本复制到自己的文件夹，以便进行编写和保存
